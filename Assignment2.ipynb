1)Area of Triangle:
Given  the  lengths  of  three  sides  of  a  triangle,  calculate  the  area  of the triangle. 

```
a = int(input("Enter the one side of triangle:"))
b = int(input("Enter the base of triangle:"))
c = int(input("Enter the side of triangle:"))
s = (a+b+c)/2
print(s)
area = (s*(s-a)*(s-b)*(s-c))**0.5
print("Area of triangle is :",area)
```

15.Write a program that finds the longest palindromic substring of a given string

```
s = input("Enter the string:")
a = s.split(" ")
c =[]
for i in a:
  b = i
  b =b[::-1]
  if i == b:
    c.append(i)
longest = max(c,key=len)
print(longest)

```
